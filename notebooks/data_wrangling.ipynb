# Data Wrangling

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
#we upload our data and explore the first five rows
file_path = '../raw_data/Mission_Hospital_pricing.xlsx' 
df = pd.read_excel(file_path, 'MH-Raw Data')
df.head()
```

```python
#We drop "SL." because it doesn't provide any info
df = df.drop('SL.', axis=1)
```

```python
#We explore the data type and info of each column
df.info()
```

The datatypes seems fine but there are some columns with null values. So, we explore each column with null value one by one.

```python
print(df['KEY COMPLAINTS -CODE'].unique())
```

```python
#We impute the null values with "unknown" because we have no context on the missing values
df['KEY COMPLAINTS -CODE'].fillna('unknown', inplace=True)
print(df['KEY COMPLAINTS -CODE'].unique())
```

```python
df[df['BP -HIGH'].isna()]
```

```python
#Since all missing BP results for children below 16 years, We assume normal BP range and we imppute the normal range 
df['BP -HIGH'].fillna(120, inplace=True)
df['BP-LOW'].fillna(80, inplace=True)
print(df['BP -HIGH'].isna().sum())
print(df['BP-LOW'].isna().sum())
```

```python
print(df['PAST MEDICAL HISTORY CODE'].unique())
```

```python
#We put "unknown" for missing past medical history code
df['PAST MEDICAL HISTORY CODE'] = df['PAST MEDICAL HISTORY CODE'].str.lower()
df['PAST MEDICAL HISTORY CODE'].fillna('unknown', inplace=True)
print(df['PAST MEDICAL HISTORY CODE'].unique())
```

```python
df[df['HB'].isna()]
```

According to https://www.ncbi.nlm.nih.gov/books/NBK259/ a normal hemoglobin level for female is 12 to 16. So we use the average of that (14) to impute for missing values

```python
#Since we don't have insight about the patients we assume a normal HB of 14
df['HB'].fillna(14, inplace=True)
df['HB'].isna().sum()
```

According to https://www.ncbi.nlm.nih.gov/books/NBK305/ the normal urea level is 5 to 20. So, we use the average of that (13) to impute for missing values 

```python
#similarly we impute normal UREA level of 13 
df['UREA'].fillna(13, inplace=True)
df['UREA'].isna().sum()
```

```python
#We impute mean value for creatinine levels 
df['CREATININE'].fillna(df['CREATININE'].mean(), inplace=True)
df.isna().sum()
```

```python
#compare categorical variable vs mean total cost
for col in df.columns:
    if df[col].dtype == 'object':
        df.groupby(by=col)['TOTAL COST TO HOSPITAL '].mean()
        plt.bar(x=df[col], height=df['TOTAL COST TO HOSPITAL '])
        plt.xticks(rotation=45)
        plt.xlabel(col)
        plt.ylabel('AVERAGE COST TO HOSPITAL(INR)' )
        plt.title(f'{col} vs TOTAL COST TO HOSPITAL(INR)')
        plt.show()
```

```python
#We make statistical summary of the numerical variables
df.describe().T
```

```python
#We make histogram of the numerical variables to see their distribution
df.hist(figsize=(15,10))
plt.subplots_adjust(hspace=0.5);
```

```python
#We expect even newborns to be average height of 49cm so any height below is suspicious. let examine height below that
df[df['BODY HEIGHT'] < 45]
```

```python
#They all seem impossible heights for the age, so we remove them
df = df[df['BODY HEIGHT'] > 45]
df.shape
```

```python
#let's see if there are other anomalies in height
plt.scatter(df['AGE'],df['BODY HEIGHT'], s=10, alpha=0.5)
plt.xlabel('AGE')
plt.ylabel('BODY HEIGHT(cm)' )
plt.title('AGE vs BODY HEIGHT')
plt.show()
```

```python
#There seem two points that require further investigation
(df[(df['AGE'] < 5) & (df['BODY HEIGHT'] > 160)])
```

```python
(df[(df['AGE'] > 35) & (df['BODY HEIGHT'] < 80)])
```

```python
#The first one seems a mistake on the age, so we impute the mean age
#The second one seems a mistake on the body height. 1 digit seems missing so impute 162 for 62
mask1 = (df['AGE'] < 5) & (df['BODY HEIGHT'] > 160)
mask2 = (df['AGE'] > 35) & (df['BODY HEIGHT'] < 80)

# Apply the conditions using loc
df.loc[mask1, 'AGE'] = df['AGE'].mean()
df.loc[mask2, 'BODY HEIGHT'] = 162
```

```python
#Check if there is similar anomaly with weight. so mak scatter plot of age vs weight
plt.scatter(df['AGE'],df['BODY WEIGHT'], s=10, alpha=0.5)
plt.xlabel('AGE')
plt.ylabel('BODY WEIGHT(Kg)' )
plt.title('AGE vs BODY WEIGHT')
plt.show()
```

```python
#clean up the column names
df.columns = df.columns.str.strip()
df.columns = df.columns.str.replace(' -', '-')
df.columns = df.columns.str.replace('- ', '-')
df.columns = df.columns.str.replace(' - ', '-')
print(df.columns)
```

```python
df.to_csv('../data/mission_hospital_cleaned.csv', index=False)
```
